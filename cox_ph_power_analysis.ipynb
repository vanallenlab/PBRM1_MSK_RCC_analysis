```python
import lifelines
import pandas as pd
```

```python
def power(sample_sizes, hr, p_exp, p_con):
    power = []
    n = []
    hrs = []
    for s in sample_sizes:
        for h in hr:
            n_exp = s*0.32
            n_con = s - n_exp
            p_exp = p_exp
            p_con = p_con
            postulated_hazard_ratio = h
            
            res = lifelines.statistics.power_under_cph(n_exp, n_con, p_exp, p_con, postulated_hazard_ratio, alpha=0.05)
            power.append(res)
            n.append(s)
            hrs.append(h)
    df = pd.DataFrame(list(zip(n, hrs, power)), 
               columns =['n_samples', 'hr', 'power'])
    return(df)
```

```python
# sample sizes for power analysis
sample_sizes = [50, 100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700,
                750, 800, 850, 900, 950, 1000]
```

```python
# postulated hazard ratios of PBRM1 status derived from 95% CI of Cox PH on entire cohort
hr = [0.5, 0.73, 1.07]
# power analysis
df = power(sample_sizes, hr, 0.65, 0.85)
```

```python
# number of experimental and control group samples needed to achieve power of 80%
# given postulated hazard ratio of 0.73
lifelines.statistics.sample_size_necessary_under_cph(0.8, 0.32,  0.65, 0.85, 0.73)
```
